```python
#Indictor Function
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

outside_rotorMag1 = outside_per_time_step[0:100000]
outside_rotorMag2 = outside_per_time_step[100000:200000]
outside_rotorMag3 = outside_per_time_step[200000:300000]
outside_attitudeMag3 = outside_per_time_step[300000:]

from statistics import mean, stdev
mean_rotorMag1, stddev_rotorMag1 = mean(outside_rotorMag1), stdev(outside_rotorMag1)
mean_rotorMag2, stddev_rotorMag2 = mean(outside_rotorMag2), stdev(outside_rotorMag2)
mean_rotorMag3, stddev_rotorMag3 = mean(outside_rotorMag3), stdev(outside_rotorMag3)
mean_attitudeMag3, stddev_attitudeMag3 = mean(outside_attitudeMag3), stdev(outside_attitudeMag3)

x1 = np.linspace(mean_rotorMag1 - 3*stddev_rotorMag1, mean_rotorMag1 + 3*stddev_rotorMag1, 100)
x2 = np.linspace(mean_rotorMag2 - 3*stddev_rotorMag2, mean_rotorMag2 + 3*stddev_rotorMag2, 100)
x3 = np.linspace(mean_rotorMag3 - 3*stddev_rotorMag3, mean_rotorMag3 + 3*stddev_rotorMag3, 100)
x4 = np.linspace(mean_attitudeMag3 - 3*stddev_attitudeMag3, mean_attitudeMag3 + 3*stddev_attitudeMag3, 100)

gaussian_rotorMag1 = norm.pdf(x1, mean_rotorMag1, stddev_rotorMag1)
gaussian_rotorMag2 = norm.pdf(x2, mean_rotorMag2, stddev_rotorMag2)
gaussian_rotorMag3 = norm.pdf(x3, mean_rotorMag3, stddev_rotorMag3)
gaussian_attitudeMag3 = norm.pdf(x4, mean_attitudeMag3, stddev_attitudeMag3)

plt.plot(x1, gaussian_rotorMag1, "y", label = "Rotor Fault Mag 1")
plt.plot(x2, gaussian_rotorMag2, "b", label = "Rotor Fault Mag 2")
plt.plot(x3, gaussian_rotorMag3, "g", label = "Rotor Fault Mag 3")
plt.plot(x4, gaussian_attitudeMag3, "r", label = "Attitude Noise - Novel")
plt.legend()
plt.show()



```

```python
from statistics import mean, stdev
vals = []
for i in range(0, len(outside_per_time_step), 1000):
  segment = outside_per_time_step[i:i+1000]
  mu = mean(segment)
  sigma = stdev(segment)
  lower_threshold = mu - 5*sigma
  upper_threshold = mu + 5*sigma
  lst = [mu, lower_threshold, upper_threshold]
  vals.append(lst)

changes = []
for i in range(1, len(vals)):
  current = vals[i]
  previous = vals[i-1]

  current_mean = current[0]
  previous_lower_thres = previous[1]
  previous_upper_thres = previous[2]

  if (current_mean > previous_upper_thres) or (current_mean < previous_lower_thres):
    start_time_step = i*1000
    end_time_step = (i+1)*1000
    changes.append(i*1000)
    #print("Change point detected between t = %s and t = %s" %(start_time_step, end_time_step))


emp = []
for i in changes:
  emp.append(outside_per_time_step[i])

x = np.arange(len(outside_per_time_step))
plt.plot(x, outside_per_time_step)
plt.scatter(changes, emp, c='red', label="change point")
plt.show()
```

```python
# taking segments of 1000, calculating the KL divergence between them
from statistics import mean, stdev
from scipy.stats import norm
def kl_divergence(p, q):
    return np.sum(np.where(p != 0, p * np.log(p / q), 0))

segments = []
kl_div_all = []

for i in range(0, len(outside_per_time_step), 1000):
  segment = outside_per_time_step[i:i+1000]
  segments.append(np.array(segment))

for i in range(1, len(segments)):
  div = kl_divergence(segments[i],segments[i-1])
  kl_div_all.append(div)
```

```python
x = np.arange(len(kl_div_all))
plt.plot(x, kl_div_all)
plt.show()
```
